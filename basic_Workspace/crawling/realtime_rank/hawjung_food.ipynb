```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
url="https://www.google.com/search?hl=ko&tbs=lf:1,lf_ui:9&tbm=lcl&q=%ED%99%94%EC%A0%95%EB%8F%99%EB%A7%9B%EC%A7%91"
response = requests.get(url)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
results = soup.select('.dbg0pd')
```

```python
for result in results:
    print(result.get_text(),"\n")
```
