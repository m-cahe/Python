### 1. url설정 -> 크롬브라우저 오픈 -> query이용하여 검색

#####   using   urllib  /  bs4  /  selenium

```python
from urllib.parse import quote_plus
```

```python
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
```

```python
baseUrl ="https://www.diningcode.com/list.php?query="
plusUrl=input("무엇을 검색할까요? : ")  #"무엇을 검색할까요? : 서울성북구\n", #{검색할 장소 입력}
url = baseUrl + quote_plus(plusUrl)

print(url)
```

```python
### quote_plus :검색어를 변환(?)시켜주는 역할
```

```python
webdriverUrl = r'C:\Users\HMIT\Downloads\chromedriver_win32\chromedriver.exe'    #cromedriver.exe 저장되어 있는 장소 및 파일 명
```

```python
driver = webdriver.Chrome(webdriverUrl)
```

```python
driver.get(url)
```

### 2. 본격 크롤링

```python
html = driver.page_source
soup = BeautifulSoup(html)
```

```python
print(soup)
```

### 식당 / 추천메뉴 / 평점 순  (보충필요: 메뉴 및 가격)

```python
btxt=soup.select('.btxt')  #select() 이용하면 list형태로 변환
print(btxt) 
for b in btxt[1:]:   #두번째 요소부터 출력
    print(b.get_text())
```

```python
stxt=soup.select('.stxt')
print(stxt)
for s in stxt[1:]:
    print(s.get_text())
```

```python
point = soup.select('.point')
print(point)
for p in point:
    print(p.get_text())
```

```python
blink = soup.select('.blink')
print(blink)
for bl in blink[1:]:
    print(bl.attrs['href'])  #상대경로로 출력 -> 절대경로로 변환 필요
```
