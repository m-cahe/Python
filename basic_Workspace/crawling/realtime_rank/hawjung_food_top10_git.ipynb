### 1. url설정 -> 크롬브라우저 오픈 -> query이용하여 검색

#####   using   urllib  /  bs4  /  selenium

```python
from urllib.parse import quote_plus
```

```python
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
```

```python
baseUrl ="https://www.diningcode.com/list.php?query="
plusUrl=input("무엇을 검색할까요? : ")  #"무엇을 검색할까요? : 서울성북구\n", #{검색할 장소 입력}
url = baseUrl + quote_plus(plusUrl)

print(url)
```

### 2. 본격 크롤링

```python
html = driver.page_source
soup = BeautifulSoup(html)
```

```python
print(soup)
```

### 식당 / 추천메뉴 / 평점 순  (보충필요: 메뉴 및 가격)

```python
btxt=soup.select('.btxt')[1:]  #select() 이용하면 list형태로 변환
print(btxt) 

btxtList = []
for b in btxt:   #두번째 요소부터 출력
    print(b.get_text().split('. ')[1])
    btxtList.append(b.get_text().split('. ')[1])  # 앞에 숫자(1. ) 없애기
    
print(btxtList)     #for문 활용하여 list형태로 바꾸고 싶다면 1)list형태 변수(a) 생성, 2) a.append사용
```

```python
stxt=soup.select('.stxt')[1:]
print(stxt)

stxtList = []

for s in stxt:
    print(s.get_text())
    stxtList.append(s.get_text())
    
print(stxtList)
```

```python
point = soup.select('.point')
print(point)

pointList = []

for p in point:
    print(p.get_text())
    pointList.append(p.get_text())

print(pointList)
```

```python
blink = soup.select('.blink')
print(blink)

blinkList = []
diningUrl = 'https://www.diningcode.com'

for bl in blink[1:]:
    print(bl.attrs['href'])  #상대경로로 출력 -> 절대경로로 변환 필요
    blinkList.append(diningUrl + bl.attrs['href'])

print(blinkList)
```

### 정리하기(DataFrame)

```python
import pandas as pd
```

```python
#List(or 다른형태)를 DataFrame형태로 변환
STORE = pd.DataFrame(btxtList)
MENU = pd.DataFrame(stxt)
POINT = pd.DataFrame(point)
LINK = pd.DataFrame(blinkList)
STORE 
MENU
POINT
LINK
```

```python
H_FOOD = pd.concat([STORE,MENU,POINT,LINK], axis=1)
#pd.DataFrame() 
#axis=1  열 기준
```

```python
H_FOOD.columns=['가게명','대표메뉴','점수','링크']
#데이터프레임 클럼명 지정 : columns : 컬럼명
```
